```python
import datetime, time
import simpy

import pandas as pd
import openclsim.core as core
import openclsim.model as model
import openclsim.plot as plot


simulation_start = 0
env = simpy.Environment(initial_time=simulation_start)
registry = {}

start = model.BasicActivity(
    ID= "start",
    env= env,
    name= "start",
    registry= registry,
    duration= 10,
    postpone_start= False,
)
```

```python


a = model.BasicActivity(
    ID= "a",
    env= env,
    name= "a",
    registry= registry,
    duration= 10,
    postpone_start= False,

)
b = model.BasicActivity(
    ID= "b",
    env= env,
    name= "b",
    registry= registry,
    duration= 10,
    postpone_start= False,
    start_event=[
        {
            "name": "a",
            "type": "activity",
            "state": "done"
        }
    ]
)
d = model.BasicActivity(
    ID= "d",
    env= env,
    name= "d",
    registry= registry,
    duration= 10,
    postpone_start= False,
    start_event=[
        {
            "name": "b",
            "type": "activity",
            "state": "done"
        }
    ]
)

c = model.BasicActivity(
    ID= "c",
    env= env,
    name= "c",
    registry= registry,
    duration= 10,
    postpone_start= False,
    start_event=[
        {
            "name": "d",
            "type": "activity",
            "state": "done"
        }
    ]
)

a.start_event=[
        {
            "or":[
                {
                    "name": "start",
                    "type": "activity",
                    "state": "done"
                },
                {
                    "name": "c",
                    "type": "activity",
                    "state": "done"
                }
            ]
        }
    ]

env.run()
```

```python
plot.vessel_planning([a,b,c,d])
```
